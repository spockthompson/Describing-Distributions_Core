```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
from scipy import stats

plt.style.use(['seaborn-muted','fivethirtyeight'])
sns.set_context('talk')
```

```python
df=pd.read_csv("Data/medical_info")
df.head()
```

```python
df.info()
```

##  VitD_levels

```python
col = 'VitD_levels'
feature = df[col]
mean = feature.mean()
median = feature.median()
std = feature.std()
plus_one_std = mean + std
minus_one_std = mean - std
fig,ax = plt.subplots(figsize=(10,6))
sns.histplot(feature ,ax=ax,stat='probability', kde = True)
ax.axvline(mean,color = 'red', linestyle='dashed', label=f'Mean = {mean:,.2f}',)
ax.axvline(median, ls='--', color = 'green', label = f'Median = {median:,.2f}')
ax.axvline(plus_one_std, color = 'black',label=f'+1 std = {plus_one_std:,.2f}')
ax.axvline(minus_one_std, color = 'black', label = f'-1 std = {minus_one_std:,.2f}')
ax.axvspan(plus_one_std, minus_one_std, color = 'yellow', zorder = 0)
ax.set_title(f'{col}')
ax.legend();
```

1. Is it Discrete or Continuous?
- The above feature is Continuous, it has multiple value possibilities. 
2. Does it have a skew? If so, which direction (+/-)
- I would say this is a normal skew, almost symmetrical.
3. What type of kurtosis does it display? (Mesokurtic, Leptokurtic, Platykurtic)
- Mesokurtic because the values cluster in the middle of the distribution evenly.
